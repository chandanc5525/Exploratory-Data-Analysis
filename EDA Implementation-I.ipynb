# How many data points for each class are present? 

```python
features = df[['bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus']]

for i in features:
    print(i)
    print(features[i].value_counts())
    print('--------------------------------')
```

# Data Visyalization Plot

```python
plt.figsize = (30,15)

plt.subplot(2,2,1)
sns.scatterplot(df.area,df.price)
plt.grid()

plt.subplot(2,2,2)
sns.scatterplot(df.bedrooms,df.price)
plt.grid()

plt.subplot(2,2,3)
sns.scatterplot(df.bathrooms,df.price)
plt.grid()

plt.subplot(2,2,4)
sns.histplot(df.price)
plt.grid()
plt. tight_layout()
plt.show()
```

# Use of FacetGrid() to plot Multiplot Graph

```python
fg = sns.FacetGrid(df,col ='bedrooms')
fg.map(sns.scatterplot,'price','area').add_legend()
plt.show()
```
